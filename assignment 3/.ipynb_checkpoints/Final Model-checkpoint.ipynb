### Load instances and their respective initial solution made through construction heuristic

```python
import json 
import pprint
import pandas as pd

pp = pprint.PrettyPrinter(indent=2)

with open('feasible_solutions.json') as f:
  test_loaded_instances = json.load(f)


#Load all instances and initial solutions into dataframe
for index, instance in enumerate(test_loaded_instances):
    print("Original List for Instance: ", index)
    pp.pprint(instance["original_list"])
    print("Solution List after Construction Heuristic for Instance: ", index)
    pp.pprint(instance["solution_list"])
    print("\n")

d = {'ifUsed' : [0,0,0], 'x-coord' : [0,0,0], 'y-coord' : [0,0,0], 'capacity' : [2,200,4], 'table space width' : [3,100,1], 'table space height' : [3,100,1]}
df = pd.DataFrame(data=d)
df
```
