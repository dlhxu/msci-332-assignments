```python
import random
import numpy as np
import pprint
```

```python
def instanceGeneration(T):
        
    #Jeffrey Quiambao's student number -> 20663172
    random.seed(3172)
    
    #instantiate list to hold instances for each period
    instances = []

    for period in range(T):
        #Generate per unit ordering cost ct Uniform[10,20]
        orderingCost = round(random.uniform(10,20),1)

        #Generate per unit holding cost ht Uniform[5,15]
        holdingCost = round(random.uniform(5,15),1)

        #Generate integer demand dt using Uniform[0,20]
        demand = round(random.uniform(0,20))

        #Set storage capacity S = 40
        storageCapacity = 40

        #Assume number of units ordered per week cannot exceed Q = 40
        maxUnitsOrdered = 40
        
        #At week 1, the retailer has 0 stock of the product, else it has stock uniformly distributed between 0 and 40
        if (period+1 == 1):
            stock = 0
        else:
            stock = "TBD"

        instances.insert(len(instances), {"orderingCost": orderingCost, "holdingCost": holdingCost, "demand":demand, 
               "storageCapacity":storageCapacity, "maxUnitsOrdered":maxUnitsOrdered, "stock": stock})
    
    return(instances)

def getStats(instances):
    orderingCost = []
    holdingCost = []
    demand = []
    storageCapacity = []
    maxUnitsOrdered = []
    #stock = []
    
    #put all generated instances into an array
    for instance in instances:
        orderingCost.insert(len(orderingCost), instance["orderingCost"])
        holdingCost.insert(len(holdingCost), instance["holdingCost"])
        demand.insert(len(demand), instance["demand"])
        storageCapacity.insert(len(storageCapacity), instance["storageCapacity"])
        maxUnitsOrdered.insert(len(maxUnitsOrdered), instance["maxUnitsOrdered"])
    
    return({"periods": len(orderingCost), "orderingCost": orderingCost, "avgOrderingCost": average(orderingCost), "stdOrderingCost": np.std(orderingCost),
          "holdingCost": holdingCost,"avgHoldingCost": average(holdingCost), "stdOrderingCost": np.std(holdingCost),
          "demand": demand, "avgDemand": average(demand), "stdDemand": np.std(demand),
          "maxUnitsOrdered": average(maxUnitsOrdered), "storageCapacity": average(storageCapacity)})

def average(lst):
    return sum(lst)/len(lst)

#Generate 10 random instances for each value T = 10, 20, 30, T weeks    
tenWeeks = instanceGeneration(10)
twentyWeeks = instanceGeneration(20)
thirtyWeeks = instanceGeneration(30)
tenWeeksStats = getStats(tenWeeks)
twentyWeeksStats = getStats(twentyWeeks)
thirtyWeeksStats = getStats(thirtyWeeks)

#pprint.pprint(tenWeeks)
print('\n')
pprint.pprint(tenWeeksStats)
print('\n')
#pprint.pprint(twentyWeeks)
print('\n')
pprint.pprint(twentyWeeksStats)
print('\n')
#pprint.pprint(thirtyWeeks)
print('\n')
pprint.pprint(thirtyWeeksStats)

#Get average and standard deviation over 10 instances
```
